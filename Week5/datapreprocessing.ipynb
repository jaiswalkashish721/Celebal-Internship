```python
import pandas as pd
df=pd.read_csv("titanic.csv")
df.head()
```

```python
df.shape
df.info()
```

```python
#data cleaning
df.isnull().sum()
```

```python
#outliers function
def outlier(df,col):
  q1=df[col].quantile(0.25)
  q3=df[col].quantile(0.75)
  iqr=q3-q1
  upperlimit=q3+1.5*iqr
  lowerlimit=q1-1.5*iqr
  return upperlimit,lowerlimit

#age outliers
upperlimit_age,lowerlimit_age=outlier(df,'Age')
print("Upper_age: ",upperlimit_age,"Lower_age: ",lowerlimit_age)
#calculate mean of non_outlier values
df_nonoutlier=df[(df['Age']>0) & (df['Age']<upperlimit_age)]
mean_age=df_nonoutlier['Age'].mean()
# replace with age>upper with upper and age<lower with lower:
for i in df['Age']:
  if i>upperlimit_age:
    df['Age']=df['Age'].replace(i,upperlimit_age)
  elif i<lowerlimit_age:
    df['Age']=df['Age'].replace(i,0)
# print(df[(df["Age"] < lowerlimit_age) | (df["Age"] > upperlimit_age)])
```

```python
#replacing null values
df['Age']=df['Age'].fillna(mean_age)
mean_fare=df['Fare'].mean()
df['Fare']=df['Fare'].fillna(mean_fare)
#sum of unique values in embarked
df['Embarked'].value_counts()
#replace nan with most freq val in embarked
df['Embarked']=df['Embarked'].replace('nan','S')
```

```python
#drop useless cols
# df.drop(['PassengerId','Survived'],axis=1,inplace=True)
df.drop(['Name','Ticket','Cabin'],axis=1,inplace=True)
df.head(10)
```

```python
#label encoding for categorical data like Embarked, Sex
from sklearn.preprocessing import LabelEncoder
le1=LabelEncoder()
df['Embarked']=le1.fit_transform(df['Embarked'])
le2=LabelEncoder()
df['Sex']=le2.fit_transform(df['Sex'])
df.head()
```
